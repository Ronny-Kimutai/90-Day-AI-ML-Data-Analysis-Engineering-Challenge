```python
#Importing Libraries - Step 1
import pandas as pd  #This is used for data handling and cleaning
import numpy as np  #This is used for numerical operations
```

```python
#Loads the CSV into a Pandas DataFrame for easy analysis - Step 2
df=pd.read_csv("../Week-1/ev_charging_patterns.csv")
```

```python
#Inspect the Data - Step 3
# What to look for include: Missing values, Data types, Basic statistics, Column names, Outliers etc.
print("First 5rows:\n", df.head())  #Prints the first 5 rows of the DataFrame
print("\nDataset Info:\n", df.info()) #Shows column names, non-null counts, and data types of each column
print("\nBasic Statistics:\n", df.describe()) #Provides summary statistics for numerical columns e.g. mean, std, min, max
print("\nColumns:\n", df.columns) #Lists all column names in the DataFrame

#NB - More can still be Included
```

```python
#Clean the Data - Step 4

#Remove Duplicates  
df = df.drop_duplicates() 

#Handle Missing Values
df = df.fillna(method='ffill')  #Forward fill (example) to propagate last valid observation

# Optional: drop rows with critical missing values
# df = df.dropna(subset=['start_time', 'end_time', 'energy_kWh'])

#NB - More cleaning steps can be added based on data understanding
# Cleaning ensures your data is consistent for usable analysis
```

```python
# Convert Timestamps - Step 5
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Charging End Time'] = pd.to_datetime(df['Charging End Time'])

#Converting to datetime alllows calculations like charging duration and time-based grouping
```
